## Problem Statement



A skier leaves the ramp of a ski jump with a velocity of $10.0$ m/s, $15.0^\circ$ above the horizontal. The slope is inclined at $50.0^\circ$, and air resistance is negligible. Youâ€™d like to know where, on the incline, the skier lands.

## Solution

Below you will find the solution to this problem.  

1. <span style="color:red">Study the code closely and with the help of the book, a classmate, or B. Nelson, figure out what each line is doing.</span>
2. <span style="color:red">Add a comment next to each line describing it.</span>

```python
from numpy import linspace,pi,cos,sin,tan,array
from numpy.linalg import norm
from matplotlib import pyplot
from sys import exit

Nskier = 100
t = linspace(0,3,Nskier)
vi = 10
theta = 15 * pi/180
vix = vi * cos(theta)
viy = vi * sin(theta)
ay = -9.8

x = vix * t
y = viy * t + 1/2 * ay * t**2

Nslope = 100
xslope = linspace(0,30,Nslope)
yslope = -tan(50*pi/180) * xslope

pyplot.plot(xslope,yslope)
pyplot.plot(x,y)
pyplot.show()


for idx in range(Nskier):
    skierPosition = array([x[idx],y[idx]])
    for index in range(Nslope):
        slopeVec = array([xslope[index],yslope[index]])
        if norm(skierPosition - slopeVec) < 1 and idx > 10:
            print("The landing location is {},{}  m".format(x[idx],y[idx]))
            exit()
```

# Problem Statement

An archer is standing on an incline when he fires his arrow at $23^\circ$ above the incline (not above the horizontal) with an initial speed of $15$ m/s.  The incline's angle is $12^\circ$.  Using plots and loops, determine where the arrow lands.




```python
from numpy import linspace,pi,cos,sin,tan,array
from numpy.linalg import norm #
from matplotlib import pyplot
from sys import exit #

Nskier = 500
t = linspace(0,1.8,Nskier)
vi = 15
theta = 35 * pi/180
vix = vi * cos(theta)
viy = vi * sin(theta)
ay = -9.8

x = vix * t
y = viy * t + 1/2 * ay * t**2

Nslope = 500
xslope = linspace(0,20,Nslope)
yslope = tan(12*pi/180) * xslope

pyplot.plot(xslope,yslope)
pyplot.plot(x,y)
pyplot.show()


def cal_average(x):
    sum_x = 0
    for t in x:
        sum_x = sum_x + t
    avg = sum_x / len(x)
    return avg

closex = []
closey = []
for idx in range(Nskier):
    skierPosition = array([x[idx],y[idx]])
    for index in range(Nslope):
        slopeVec = array([xslope[index],yslope[index]])
        if norm(skierPosition - slopeVec) < 1 and idx > 100:
            closex.append(x[idx])
            closey.append(y[idx])
            #print("The landing location is {},{}  m".format(x[idx],y[idx]))
print(f'The landing location is ({cal_average(closex):.2f}, {cal_average(closey):.2f})')
```
